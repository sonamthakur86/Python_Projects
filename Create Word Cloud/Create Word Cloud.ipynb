```python
import wikipedia
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
import re
```

```python
text=wikipedia.page("Python Programming Language") # Mension the heading of wikipedia page
```

```python
content=text.content # Store the whole content into content variable
```

```python
print(content) 
```

```python
content=re.sub(r'==.*?==+','',content) # Remove all the symbols from content
```

```python
content=content.replace("\n",'') # Remove spaces
```

```python
print(content)
```

```python
stopwords=set(STOPWORDS) #Collect stopwords and remove them
```

```python
print(stopwords)
```

```python
wordcloud=WordCloud(width=800,height=800,background_color='white',stopwords=stopwords,min_font_size=10).generate(content)
```

```python
plt.figure(figsize=(8,8),facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.show() #Print wordcloud using matplotlib
```
